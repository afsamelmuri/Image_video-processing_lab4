```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread(r"C:\Users\Lenovo\Downloads\daimond pp\2nd sem video lab\football.jfif", 0)  # Add your image path here

# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(image)

# Display original and equalized images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Equalized Image')
plt.imshow(equalized_image, cmap='gray')

plt.show()
```

```python
def contrast_stretching(image):
    # Min-max stretching
    a, b = np.min(image), np.max(image)
    stretched_image = (image - a) * (255 / (b - a))
    return np.uint8(stretched_image)

# Load the image in grayscale
image = cv2.imread(r"C:\Users\Lenovo\Downloads\daimond pp\2nd sem video lab\football.jfif", 0)  # Add your image path here

# Apply contrast stretching
stretched_image = contrast_stretching(image)

# Display the result
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Contrast Stretched Image')
plt.imshow(stretched_image, cmap='gray')

plt.show()
```

```python
def gamma_correction(image, gamma):
    # Build a lookup table mapping the pixel values [0, 255] to their adjusted gamma values
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")

    # Apply gamma correction using the lookup table
    return cv2.LUT(image, table)

# Load the image in grayscale
image = cv2.imread(r"C:\Users\Lenovo\Downloads\daimond pp\2nd sem video lab\football.jfif", 0)  # Add your image path here

# Apply gamma correction with gamma = 2.0
gamma_corrected = gamma_correction(image, 2.0)

# Display the result
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Gamma Corrected Image')
plt.imshow(gamma_corrected, cmap='gray')

plt.show()
```

```python
def logarithmic_transformation(image):
    # Apply logarithmic transformation
    c = 255 / np.log(1 + np.max(image))
    log_image = c * (np.log(1 + image))
    return np.uint8(log_image)

# Load the image in grayscale
image = cv2.imread(r"C:\Users\Lenovo\Downloads\daimond pp\2nd sem video lab\football.jfif", 0)  # Add your image path here

# Apply logarithmic transformation
log_image = logarithmic_transformation(image)

# Display the result
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Logarithmic Transformed Image')
plt.imshow(log_image, cmap='gray')

plt.show()
```

```python
def unsharp_masking(image):
    # Gaussian blur
    blurred_image = cv2.GaussianBlur(image, (9, 9), 10.0)
    
    # Unsharp masking
    sharpened_image = cv2.addWeighted(image, 1.5, blurred_image, -0.5, 0)
    return sharpened_image

# Load the image in grayscale
image = cv2.imread(r"C:\Users\Lenovo\Downloads\daimond pp\2nd sem video lab\football.jfif", 0)  # Add your image path here

# Apply unsharp masking
sharpened_image = unsharp_masking(image)

# Display the result
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Sharpened Image')
plt.imshow(sharpened_image, cmap='gray')

plt.show()
```
